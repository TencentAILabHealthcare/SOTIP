```python
polar_join = adata_sotip.var[['sotip_polar_to_tumor_bh','gt_polarized']]
```

```python
adata_spagcn.var
```

```python
polar_join['spagcn_polar_to_tumor_bh'] = adata_spagcn.var['spagcn_polar_to_tumor_bh']
```

```python
fdr_array = np.arange(0,0.5,0.01)
pd_dict = {
    'fdr':[],
    'power':[],
    'method':[]
}
gt_polar_count = np.sum(polar_join['gt_polarized'])
for fdr in fdr_array:
    sotip_polar_count = np.sum(polar_join[polar_join['sotip_polar_to_tumor_bh']<=fdr]['gt_polarized'])
    spagcn_polar_count = np.sum(polar_join[polar_join['spagcn_polar_to_tumor_bh']<=fdr]['gt_polarized'])  
    
    sotip_power = sotip_polar_count/gt_polar_count
    spagcn_power = spagcn_polar_count/gt_polar_count
    pd_dict['fdr'].append(fdr)
    pd_dict['power'].append(sotip_power)    
    pd_dict['method'].append('sotip')    
    
    pd_dict['fdr'].append(fdr)
    pd_dict['power'].append(spagcn_power)    
    pd_dict['method'].append('spagcn')    
pd_df = pd.DataFrame(pd_dict)
```

```python
cmp = palettable.cartocolors.qualitative.Bold_10.mpl_colors[0:2]
g = sns.lineplot(data=pd_df,x='fdr',y='power',hue='method',palette=cmp)
g.legend_.remove()
plt.title('scMEP_p23')
```
